### Model building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
employee_data
```

```python
#  One Hot encoding
# Label Encoder 
```

```python
employee_data['dept'].value_counts()
```

```python
employee_data['salary'].value_counts()
```

```python
emp = pd.get_dummies(data=employee_data, columns=['dept'],drop_first=True)
```

```python
emp
```

```python
emp['salary'] = emp['salary'].apply(lambda x :['low','medium','high'].index(x))
```

```python
emp['salary'].value_counts()
```

```python
emp
```

```python
employee_data.head()
```

```python
emp.head()
```

```python
employee_data['Attrition'].value_counts(normalize=True)*100
```

* we have a moderately imbalanced data set

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
employee_data[['Emp ID','Attrition']]
```

```python
emp.head()
```

```python
x = emp.drop(['Quit', 'Attrition', 'Emp ID'], axis=1)
y = emp['Attrition']
```

```python
x
```

```python
y.value_counts()
```

```python
x_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
x_resampled
```

```python
y_resampled.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size= 0.3, random_state=1, shuffle=True)
```

```python
x_train
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
lr.score(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
print('\n')
print(accuracy_score(y_test, y_pred))
print('\n')
print(confusion_matrix(y_test, y_pred))
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
dt.score(x_train, y_train)
```

```python
y_pred1 = dt.predict(x_test)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test, y_test)
```

```python
rf.score(x_train, y_train)
```

```python
y_pred2 = rf.predict(x_test)
```
